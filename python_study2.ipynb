```python
xmlpath=np.array(glob("/home/alex/prog/jobs/1904/check0401/*.xml"))
xmlpath
```

```python
xmlpath[0]
```

```python
def xmltojson(xmlstr):
    #parse是的xml解析器
    xmlparse = xmltodict.parse(xmlstr)
    #json库dumps()是将dict转化成json格式，loads()是将json转化成dict格式。
    #dumps()方法的ident=1，格式化json
    jsonstr = json.dumps(xmlparse,indent=1)
    return jsonstr
```

```python
lines = open(xmlpath[0]).readlines()
lines
```

```python
lines="".join(lines)
print(lines)
```

```python
jsonStr=xmltojson(lines)
```

```python


with open("./xml2json.json",'w+') as f:
    for o in jsonStr:
        f.write(o)
```

### opencv读png转成jpg

```python

import os
import cv2
import sys
import numpy as np
 
path = "/home/alex/prog/jobs/1904/heatmap/image_stitching/pic4/"
print(path)
 
for filename in os.listdir(path):
    if os.path.splitext(filename)[1] == '.jpg':
        # print(filename)
        img = cv2.imread(path + filename)
        print(filename.replace(".jpg",".png"))
        newfilename = filename.replace(".jpg",".png")
        # cv2.imshow("Image",img)
        # cv2.waitKey(0)
        
        cv2.imwrite(path + newfilename,img)

```

```python
for filename in os.listdir(path):
    if os.path.splitext(filename)[1] == '.png':
        img = cv2.imread(path + filename)
        img=cv2.resize(img,(1280,720),interpolation=cv2.INTER_AREA)
        newfilename = 'new'+filename
        cv2.imwrite(path + newfilename,img)
```
